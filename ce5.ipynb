```python
import geopandas as gpd
from cartopy import crs as ccrs
```

```python
usa = gpd.read_file(r"C:\Users\andre\OneDrive\Documents\schoolstuff\AES408\states\tl_2025_us_state.shp")
pop = gpd.read_file(r"C:\Users\andre\OneDrive\Documents\schoolstuff\AES408\populated_places\ne_10m_populated_places.shp")
usa = usa.to_crs("EPSG:5070")
pop = pop.to_crs("EPSG:5070")
```

```python
def pop_places(usa,pop):
    result = gpd.sjoin(usa,pop,predicate='within')
    alabama = usa[usa['NAME'] == 'Alabama']
    return result
```

```python
alabama_places = pop_places(pop,alabama)
print(alabama_places['NAME_left']) # NAME_left is the column with city names from the spatial join result
```
