```python
def open_shapefile(x):
    return gpd.read_file(x)
open_shapefile(r"C:\Users\asanders\Documents\usa_shapefile\tl_2025_us_state.shp")
```

```python
def open_shapeful_al(x):
    full_fil = gpd.read_file(x)
    return full_fil[full_fil['STUSPS'] == 'AL']
open_shapeful_al(r"C:\Users\asanders\Documents\usa_shapefile\tl_2025_us_state.shp")
```
